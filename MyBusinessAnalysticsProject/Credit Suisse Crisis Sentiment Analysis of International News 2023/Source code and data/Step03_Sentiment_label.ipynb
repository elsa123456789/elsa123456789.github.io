```python
import pandas as pd
from transformers import pipeline
from textblob import TextBlob

# 加载预训练的 ChatGPT 模型
# model = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B', device=0)
# 读取Excel文件并将标题存储在一个列表中
df = pd.read_excel('processed data.xlsx')
titles = df['headline_text'].tolist()

# 定义函数，使用 TextBlob 进行情感分析
def label_attitude(title):
    blob = TextBlob(title)
    sentiment = blob.sentiment.polarity
    return sentiment


# 对每个标题进行情感分析，并将结果添加到新的列中
attitudes = []
for title in titles:
    attitude = label_attitude(title)
    attitudes.append(attitude)

df['attitude'] = attitudes

# 保存带有标记结果的 Excel 文件
df.to_excel('news_titles_with_attitude.xlsx', index=False)
```
